```python
import cv2
import matplotlib.pyplot as plt
from IPython import display
import numpy as np
import copy
```

```python
vc = cv2.VideoCapture(0)

if vc.isOpened():
    is_capturing, frame = vc.read()
#     frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    oldframe = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    webcam_preview = plt.imshow(frame)   
else:
    is_capturing = False

# plt.draw()
# display.display(plt.gcf())
# vc.release()
```

```python
while is_capturing:
    try:
        is_capturing, frame = vc.read()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        webcam_preview.set_data(frame)
        plt.draw()
        b = frame[50, :, :]
        display.clear_output(wait=True)
        display.display(plt.gcf())
        oldframe = frame
        plt.pause(0.1)
    except KeyboardInterrupt:
        break
#         vc.release()
```

```python
vc.release()
```

```python
def writeImg(frame, name):
    cv2.imwrite(name+'.jpg', cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))
```

```python
backup = copy.deepcopy(frame)
# frame = copy.deepcopy(backup)
```

```python
writeImg(frame, 'old')
def encryptFrame(frame, x, y):
    frame[x[0]:x[1], y[0]:y[1], :] = 0
    return frame

x = (50, 400)
y = (100, 550)

editframe = encryptFrame(frame, x, y)
plt.imshow(editframe)
display.display(plt.gcf())
writeImg(editframe, 'new')
```
